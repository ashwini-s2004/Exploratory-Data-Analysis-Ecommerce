###Product Distribution

```python
df['Product'].value_counts()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    data=df,
    x="Product",
    order=df["Product"].value_counts().index
)

plt.xticks(rotation=45)

plt.title("Products Sold")

plt.show()
```

Printer is the most sold product.

###Monthly Sales Trend

```python
df["Month"] = df["Date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("Month")["TotalPrice"].sum()

monthly_sales.plot(figsize=(12,5))

plt.title("Monthly Sales")

plt.show()
```

July is the highest sales month .
April is the lowest sales month .

###Product-wise Revenue

```python
from itertools import product
product_revenue = df.groupby("Product")["TotalPrice"].sum()

product_revenue.plot(kind="bar", figsize=(10,5))

plt.title("Product-wise Revenue")

plt.show()
```

###Correlation Matrix

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.select_dtypes(include=[np.number]).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```
